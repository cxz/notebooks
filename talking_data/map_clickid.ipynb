```python
""" Map click_id from old test to new test.csv
"""

import os
import pandas as pd
```

```python
!ls ../input
```

```python
DTYPES = {
    'ip'            : 'uint32',
    'app'           : 'uint16',
    'device'        : 'uint16',
    'os'            : 'uint16',
    'channel'       : 'uint16',
    'is_attributed' : 'uint8',
    'click_id'      : 'uint32'
}

test = pd.read_csv('../input/test.csv', dtype=DTYPES, parse_dates=['click_time'])
test_v0 = pd.read_csv('../input/test_v0.csv', dtype=DTYPES, parse_dates=['click_time'])
```

```python
test.set_index(['ip', 'app', 'device', 'os', 'channel', 'click_time'], inplace=True)
```

```python
test_v0 = test_v0.rename(columns={'click_id': 'click_id_v0'})
test_v0.set_index(['ip', 'app', 'device', 'os', 'channel', 'click_time'], inplace=True)
```

```python
out = test.join(test_v0, how='left')
```

```python
out = out.sort_values(by='click_id')
```

```python
out[['click_id', 'click_id_v0']].to_csv('test_mapping.csv', index=False)
```
