```python
from collections import Counter

import pandas as pd
import numpy as np

from sklearn import preprocessing
```

```python
train = pd.read_csv('input/train.csv')
train_rows = len(train)
print(train_rows)
```

```python
print('loading test')
test = pd.read_csv('input/test.csv')
```

```python
def most_common_words(df, c, most_common):
    counts = dict(Counter(df[c]).most_common(most_common))
    return counts
    #words = counts.keys()
    #title = df[c]
    #title2 = title.copy()
    #title2.loc[~title.isin(words)] = ""
    #return title2

counts = most_common_words(train[~train.description.isnull()], 'description', 10000)
```

```python
from tqdm import tqdm

out = []

for idx, (desc, count) in tqdm(enumerate(sorted(counts.items(), key=lambda x: x[1], reverse=True))):
    
    # items with exactly same description
    items = train[train.description==desc]    
    assert len(items) == count
    
    deal_max = np.max(items.deal_probability)
    deal_mean = np.mean(items.deal_probability)
    deal_median = np.median(items.deal_probability)    
    count = len(items)
    
    out.append([idx, desc, count, deal_mean, deal_median, deal_max])
    
common_descs = pd.DataFrame(out, columns=['idx', 'desc', 'count', 'deal_mean', 'deal_median', 'deal_max'])
    
```
